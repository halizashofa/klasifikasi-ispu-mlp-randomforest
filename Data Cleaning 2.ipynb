```python
import pandas as pd
import numpy as np
```

```python
# Import data

df1 = pd.read_csv('drive/My Drive/data set ISPU 2021/januari 2021.csv')
df2 = pd.read_csv('drive/My Drive/data set ISPU 2021/februari 2021.csv')
df3 = pd.read_csv('drive/My Drive/data set ISPU 2021/maret 2021.csv')
df4 = pd.read_csv('drive/My Drive/data set ISPU 2021/april 2021.csv')
df5 = pd.read_csv('drive/My Drive/data set ISPU 2021/mei 2021.csv')
df6 = pd.read_csv('drive/My Drive/data set ISPU 2021/juni 2021.csv')
df7 = pd.read_csv('drive/My Drive/data set ISPU 2021/juli 2021.csv')
df8 = pd.read_csv('drive/My Drive/data set ISPU 2021/agustus 2021.csv')
df9 = pd.read_csv('drive/My Drive/data set ISPU 2021/september 2021.csv')
df10 = pd.read_csv('drive/My Drive/data set ISPU 2021/oktober 2021.csv')
df11 = pd.read_csv('drive/My Drive/data set ISPU 2021/november 2021.csv')
df12 = pd.read_csv('drive/My Drive/data set ISPU 2021/desember 2021.csv')
```

```python
# Cek panjang data

print(df1.shape)
print(df2.shape)
print(df3.shape)
print(df4.shape)
print(df5.shape)
print(df6.shape)
print(df7.shape)
print(df8.shape)
print(df9.shape)
print(df10.shape)
print(df11.shape)
print(df12.shape)
```

```python
df1
```

```python
# Gabungkan data

df_2021 = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12]).reset_index().drop(columns=['index'])
df_2021
```

```python
# Export file data ISPU 2021

df_2021.to_excel('Data ISPU 2021.xlsx', index=False)
```

```python
df_2021.info()
```

```python
# Cek missing value

df_2021.isnull().sum()
```

```python
# Cek kelas data

df_2021.categori.unique()
```

```python
df_2021_new = df_2021.copy()
```

```python
# Buang string

for kolom in df_2021_new.drop(columns=['tanggal','stasiun','critical']).columns:
  df_2021_new[kolom] = df_2021_new[kolom].replace(r'---',np.NaN,regex=True)

df_2021_new
```

```python
df_2021_new.info()
```

```python
df_2021_new.isna().sum()
```

```python
# Buang NaN

df_2021_new = df_2021_new.dropna(axis=0)
df_2021_new.isnull().sum()
```

```python
# Konversi str menjadi int

df_2021_new = df_2021_new.astype({'pm10': int, 'pm25': int, 'so2': int, 'co': int, 'o3': int, 'no2': int, 'max': int})

df_2021_new.info()
```

```python
# Export file data ISPU 2021 bersih

df_2021_new.to_excel('ISPU 2021 Hasil Cleaning.xlsx', index=False)
```
